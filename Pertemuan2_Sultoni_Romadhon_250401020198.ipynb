```python
import numpy as np

# Random generator
rng = np.random.default_rng(seed=42)

# Operasi aritmetika vektorisasi
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print(a + b)   # [11 22 33 44 55]
print(a * b)   # [10 40 90 160 250]
print(a ** 2)  # [1 4 9 16 25]

# Universal functions (ufuncs)
print(np.sqrt(a))
print(np.exp(a))
print(np.log(b))

# Fungsi statistik
data = rng.normal(0, 1, 1000)

print(f'Mean: {np.mean(data):.4f}')
print(f'Std: {np.std(data):.4f}')
print(f'Min: {np.min(data):.4f}')
print(f'Max: {np.max(data):.4f}')
print(f'Q1: {np.percentile(data,25):.4f}')
print(f'Q3: {np.percentile(data,75):.4f}')
```

```python
import pandas as pd

# Dataset Titanic (contoh URL publik)
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(titanic_url)

# Filter satu kondisi
penumpang_selamat = df[df['Survived'] == 1]
penumpang_wanita = df[df['Sex'] == 'female']

# Filter kondisi ganda (& = AND, | = OR)
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

# Seleksi kolom tertentu
kolom_terpilih = df[['Name', 'Age', 'Sex', 'Survived']]

# Output
print(penumpang_selamat.head())
print(penumpang_wanita.head())
print(wanita_kelas1.head())
print(kolom_terpilih.head())
```

```python
# Langkah 4: Impor library
import pandas as pd


# Langkah 5: Muat dataset Titanic
# Catatan: Gunakan link raw agar data terbaca sempurna oleh Pandas
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Cek 5 baris pertama untuk memastikan data berhasil dimuat
df.head()
```

```python
# Q1 Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?
# Jawaban Q1
print(f"Jumlah baris: {df.shape[0]}")
print(f"Daftar kolom: {df.columns.tolist()}")
```

```python
# Q2 Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)
# Jawaban Q2
missing = df.isnull().sum()
persen = (missing / len(df)) * 100

print("Jumlah missing per kolom:")
print(missing)

print("\nPersentase missing (%):")
print(persen)
```

```python
# Q3 Berapa persen penumpang yang selamat (Survived=1)?
# Jawaban Q3
print(f"Persentase selamat: {df['Survived'].mean() * 100:.2f}%")
```

```python
# Q4 Filter penumpang wanita dari kelas 1 (Pclass=1)
# Jawaban Q4
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

print(f"Jumlah wanita kelas 1: {len(wanita_kelas1)}")
print(f"Rata-rata usia: {wanita_kelas1['Age'].mean():.2f}")
```

```python
# Q5 Tingkat keselamatan per kelas penumpang
# Jawaban Q5
survival_per_class = df.groupby('Pclass')['Survived'].mean()

print("Survival rate per kelas:")
print(survival_per_class)

print(f"Kelas dengan survival tertinggi: {survival_per_class.idxmax()}")
```

```python
# Q6 Analisis tambahan (berdasarkan gender)
# Jawaban Q6
survival_gender = df.groupby('Sex')['Survived'].mean()

print("Survival rate berdasarkan gender:")
print(survival_gender)

print("\nInterpretasi: Wanita memiliki peluang selamat lebih tinggi dibanding pria.")
```
